**FILTRO 1**
Se selecciono aquellos datos que en la columna "region" tenga Caribbean

**FILTRO 2**
Se selecciono los datos que esten en la fila 100 200 y 230

```python
filtro2=df.iloc[[100,200,230] , :]
filtro2
```

```python
filtro2.to_csv("Filtro2.csv")
```

**FILTRO 3: Filtro por columnas**
Se selecciono que todos los datos nada mas arrojen las columnas 1 3 y 5

```python
filtro3=df.iloc[:, [1,3,5]]
filtro3
```

```python
filtro3.to_csv("Filtro3.csv")
```

**FILTRO 4: Filtro de fila vs columna**
Se selecciono que solo aparezcan los datos que sean "Middle East"y"Eastern Africa" y que solo apareza la columna "code" de cada uno de estos.

```python
df=pd.read_csv("CountryTable.csv", index_col=3)
df
```

```python
filtro4=df.loc[["Middle East","Eastern Africa"],["code"]]
filtro4
```

```python
filtro4.to_csv("Filtro4.csv")
```

**FILTRO 5: Cabecera**
Se selecciono que solo aparezcan los 10 primeros datos

```python
filtro5=df.head(10)
filtro5
```

```python
filtro5.to_csv("Filtro5.csv")
```

**FILTRO 6: Por comparacion**
Se selecciono que solo aparezcan las filas en las cuales la columna "life_expectancy" sea mayor a 77

```python
filtro6=df[df["life_expectancy"] > 77]
filtro6
```

```python
filtro6.to_csv("Filtro6.csv")
```

**FILTRO 7: AND**
Se selecciono las filas que cumplan que "life_expectancy" sea mayor a 77 y que "continent" sea "Europe"

```python
filtro7=df[(df["life_expectancy"]> 77) & (df["continent"]=="Europe")]
filtro7
```

```python
filtro7.to_csv("Filtro7.csv")
```

**FILTRO 8: OR**
Se selecciono las filas que cumplan que "life_expectancy" sea mayor a 77 o que "continent" sea "Europe"

```python
filtro8=df[(df["life_expectancy"]> 77) | (df["continent"]=="Europe")]
filtro8
```

```python
filtro8.to_csv("Filtro8.csv")
```

**FILTRO 9: NOT**
Se selecciono todas las filas que no cumplan que "life_expectancy" sea mayor a 77 y que "continent" sea "Europe"

```python
filtro9=df[~ (df["life_expectancy"]> 77) & ~ (df["continent"]=="Europe")]
filtro9
```

```python
filtro9.to_csv("Filtro9.csv")
```
